#### Edge Detection

##### Sobel operator

Derivative in X direction

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Sobel x")
# original_image,ddepth,xorder,yorder,kernelsize
sobel = cv2.Sobel(src = reimage, ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 3)
cv2_imshow(sobel)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Sobel x")
# original_image,ddepth,xorder,yorder,kernelsize
sobelx = cv2.Sobel(src = grayimg, ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 3)
cv2_imshow(sobelx)
```

Derivative in Y direction

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Sobel y")
# original_image,ddepth,xorder,yorder,kernelsize
sobel = cv2.Sobel(src = reimage, ddepth = cv2.CV_64F, dx = 0, dy = 1, ksize = 3)
cv2_imshow(sobel)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Sobel y")
# original_image,ddepth,xorder,yorder,kernelsize
sobely = cv2.Sobel(src = grayimg, ddepth = cv2.CV_64F, dx = 0, dy = 1, ksize = 3)
cv2_imshow(sobely)
```

Derivatives in X and Y direction

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Sobel x y")
# original_image,ddepth,xorder,yorder,kernelsize
# ddepth = cv2.CV_64F means higher precision so as to hold both positive and negative derivatives
sobel = cv2.Sobel(src = reimage, ddepth = cv2.CV_64F, dx = 1, dy = 1, ksize = 3)
cv2_imshow(sobel)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Sobel x y")
# original_image,ddepth,xorder,yorder,kernelsize
sobel = cv2.Sobel(src = grayimg, ddepth = cv2.CV_64F, dx = 1, dy = 1, ksize = 3)
cv2_imshow(sobel)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Sobel x y (not using cv2 function)")
# simply adding the output images for the sobel x and y derivatives
cv2_imshow(sobelx + sobely)
```

##### Prewitt operator

```python
kernelY = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernelX = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
kernelX
```

```python
kernelY
```

Derivative in X direction

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Prewitt x")
# original_image,ddepth,kernel
# ddepth = -1 means the precision should be the same as source image
# for convolving prewitt filter
prewittx = cv2.filter2D(reimage, -1, kernelX)
cv2_imshow(prewittx)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Prewitt x")
# original_image,ddepth,kernel
prewittgx = cv2.filter2D(grayimg, -1, kernelX)
cv2_imshow(prewittgx)
```

Derivative in Y direction

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Prewitt y")
# original_image,ddepth,kernel
prewitty = cv2.filter2D(reimage, -1, kernelY)
cv2_imshow(prewitty)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Prewitt y")
# original_image,ddepth,kernel
prewittgy = cv2.filter2D(grayimg, -1, kernelY)
cv2_imshow(prewittgy)
```

Derivative in X and Y directions

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Prewitt x y")
cv2_imshow(prewittx + prewitty)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Prewitt x y")
cv2_imshow(prewittgx + prewittgy)
```

##### Laplacian filter

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Laplacian")
# original_image,ddepth,kernelsize
laplace = cv2.Laplacian(src = reimage, ddepth = cv2.CV_64F, ksize = 3)
cv2_imshow(laplace)
```

```python
print("Original RGB image")
cv2_imshow(reimage)
print("Laplacian (without gaussian noise)")
# original_image,ddepth,kernelsize
denoised = cv2.GaussianBlur(reimage, (3,3), 0)
laplace = cv2.Laplacian(src = denoised, ddepth = cv2.CV_64F, ksize = 3)
cv2_imshow(laplace)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Laplacian")
# original_image,ddepth,kernelsize
laplace = cv2.Laplacian(src = grayimg, ddepth = cv2.CV_64F, ksize = 3)
cv2_imshow(laplace)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Laplacian (without gaussian noise)")
denoised = cv2.GaussianBlur(grayimg, (3,3), 0)
# original_image,ddepth,kernelsize
laplace = cv2.Laplacian(src = denoised, ddepth = cv2.CV_64F, ksize = 3)
cv2_imshow(laplace)
```

#### Observation
1. Sobel operator works well in detecting the edges, they aren't sharp but blurred. It is observed that combining the x and y derivatives of Sobel using Opencv gives a poor output, however manually adding the derivatives gives a better result.
2. Prewitt operator works similar to the sobel operator, but detects lesser number of edges probably because it gives lesser importance to the pixels closer to the center pixel, than sobel. It is observed that combining the x and y derivatives of Prewitt gives a noisy image.
3. The Laplacian works the worst probably because it is sensitive to noise. However, a better result is obtained by removing gaussian noise.
4. Canny works the best for this medical image, and detects all the sharp edges. However upon varying the thresholds, it is observed that a lower threshold value introduces a lot of noise in the output, and a higher threshold value ignores important edges (both having the same range = 100). Also, if the lower threshold value is lower (for range = 150), there is a lot of noise observed, and if the lower threshold value is higher, the noise is lesser, but some edges are ignored (for range = 50). Hence, the values lower = 100 and upper = 200 work the best for this image. Also, it is noteworthy that after manually applying Gaussian Blur and removing gaussian noise, the results of Canny filter are much more refined.
