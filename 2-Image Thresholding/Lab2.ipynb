```python
print("Original RGB image")
cv2_imshow(reimage)
print("To Zero Inverted thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(reimage, 120, 255, cv2.THRESH_TOZERO_INV)
cv2_imshow(thresh)
```

##### 2. Adaptive

- It works only on Greyscale images
- The various parameters of Adaptive thresholding:
1. cv2.ADAPTIVE_THRESH_MEAN_C
2. cv2.ADAPTIVE_THRESH_GAUSSIAN_C

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Adaptive Mean")
# source image, maxVal, adaptiveMethod, thresholdType, blocksize, constant
thresh = cv2.adaptiveThreshold(grayimg, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                          cv2.THRESH_BINARY, 11, 6)
cv2_imshow(thresh)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Adaptive Gaussian")
# source image, maxVal, adaptiveMethod, thresholdType, blocksize, constant
thresh = cv2.adaptiveThreshold(grayimg, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 5)
cv2_imshow(thresh)
```

##### 3.Otsu

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Otsu Binarization + Binary Thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(grayimg, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
cv2_imshow(thresh)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Otsu Binarization + Inverted Binary Thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(grayimg, 120, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU) 
cv2_imshow(thresh)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Otsu Binarization + Truncated thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(grayimg, 120, 255, cv2.THRESH_TRUNC + cv2.THRESH_OTSU) 
cv2_imshow(thresh)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Otsu Binarization + To Zero Thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(grayimg, 120, 255, cv2.THRESH_TOZERO + cv2.THRESH_OTSU) 
cv2_imshow(thresh)
```

```python
print("Original Gray image")
cv2_imshow(grayimg)
print("Otsu Binarization + To Zero Inverted Thresh")
# source image, thresholdValue, maxVal, technique
ret, thresh = cv2.threshold(grayimg, 120, 255, cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU) 
cv2_imshow(thresh)
```

#### Observation
1. In simple thresholding, a single threshold value is applied throughout the image. By varying the parameters for the thresholding type, we get different variations of the thresholded binary image for the grayscale image. For the color image also we observe the pixel values above the threshold are displayed for the binary threshold parameter, and similarly for the other parameters the necessary functions are applied.
2. In adaptive thresholding, using block size as 11, the mean and gaussian functions are applied, on grayscale images. It doesn't work on color images. However a lot of noise is observed and all the foreground objects are not separated out.
3. The Otsu Binarization method is applied using the normal simple thresholding method but with THRESH_OTSU as the parameter by combining with the different simple thresholding methods. This also works on grayscale images but its able to separate out most of the foreground objects with the least amount of noise. This method works the best out of all the methods.
